```python
# ============================================================
# TEST SINGLE IMAGE
# ============================================================
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

model = tf.keras.models.load_model("flower_cnn_model.h5")

class_names = [
    "daisy",
    "dandelion",
    "rose",
    "sunflower",
    "tulip"
]
# ============================================================
# CHANGE IMAGE PATH HERE
# ============================================================

img_path = "/content/tải xuống (2).jpg"
# ============================================================
# LOAD IMAGE
# ============================================================

# Ensure IMG_SIZE is defined if not already in the global scope
# From previous cells, IMG_SIZE was defined as 200.
IMG_SIZE = 200

img = tf.keras.preprocessing.image.load_img(

    img_path,

    target_size=(IMG_SIZE, IMG_SIZE)
)

img_array = tf.keras.preprocessing.image.img_to_array(img)

img_array = img_array / 255.0

img_array = np.expand_dims(img_array, axis=0)


# ============================================================
# PREDICT
# ============================================================

prediction = model.predict(img_array)

predicted_index = np.argmax(prediction)

predicted_class = class_names[predicted_index]

confidence = prediction[0][predicted_index] * 100


# ============================================================
# SHOW RESULT IMAGE
# ============================================================

# Need to import matplotlib.pyplot as plt and numpy as np if not already in global scope

plt.figure(figsize=(6,6))

plt.imshow(tf.keras.preprocessing.image.load_img(img_path))

plt.axis("off")

plt.title(
    f"{predicted_class} ({confidence:.2f}%)"
)

plt.show()


# ============================================================
# PRINT RESULT
# ============================================================

print("\n========== RESULT ==========")

print(f"Flower: {predicted_class}")

print(f"Confidence: {confidence:.2f}%")
```
