```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import img_to_array

# 1. LOAD MODEL VÀ ĐỊNH NGHĨA NHÃN THỦ CÔNG
# Đọc mô hình bạn đã train và tải lên
model = load_model('/content/mo_hinh_mon_an.keras')

# Tự định nghĩa nhãn đã train
labels = {
    0: "Bánh bèo",
    1: "Bánh cuốn",
    2: "Bánh mì",
    3: "Bánh xèo",
    4: "Cá kho",
    5: "Cơm tấm"
    }
```

```python
# 2. ĐƯỜNG DẪN ẢNH TEST
image_path = "/content/banhbeo.jpg"

# 3. ĐỌC ẢNH
image = cv2.imread(image_path)
# OpenCV mặc định đọc ảnh hệ BGR
rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# 4. TIỀN XỬ LÝ ẢNH
# Model yêu cầu ảnh kích thước 200x200
resized_img = cv2.resize(rgb, (200, 200))
# Chuẩn hóa về [0, 1]
img_normalized = resized_img.astype("float") / 255.0
# Chuyển thành array
img_array = img_to_array(img_normalized)
# Thêm trục batch_size để thành (1, 200, 200, 3) cho Keras
img_expanded = np.expand_dims(img_array, axis=0)

# 5. PREDICT (DỰ ĐOÁN)
prediction = model.predict(img_expanded)
# Lấy ra vị trí có xác suất cao nhất
predicted_class = int(np.argmax(prediction))
confidence = np.max(prediction)
# Lấy tên món ăn từ dictionary labels
money_name = labels.get(predicted_class, f"Món ăn chưa xác định (Class_{predicted_class})")

# 6. HIỂN THỊ KẾT QUẢ
print("======================================")
print("🎯 KẾT QUẢ NHẬN DIỆN MÓN ĂN")
print("======================================")
print(f"Món ăn dự đoán: {money_name}")
print(f"Độ tin cậy: {round(confidence * 100, 2)} %")

# Hiển thị ảnh
plt.figure(figsize=(10, 10))
plt.imshow(rgb)
plt.axis("off")
plt.show()
```
